```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('device_collected.csv')
columns=data.columns
consider=['timestamp','usingPhone']

x=[]
y=[]
for name in columns:
    if name in consider and name==consider[0]:
        timestamp_col=np.asarray(data[name])
        x.append(np.asarray(data[name]))
    elif name in consider and name==consider[1]:
        y.append(np.asarray(data[name]))
    else:
        pass
```

```python
days={}

timestamps=x[0]
device_label=y[0]

for i in range(len(timestamps)):
    date_time=timestamps[i].split(' ')
    if date_time[0] not in days:
        temp_dict={}
        temp_dict[date_time[1].split(':')[0]]=[device_label[i]]
        days[date_time[0]]=temp_dict
    else:
        if date_time[1].split(':')[0] not in days[date_time[0]]:
            days[date_time[0]][date_time[1].split(':')[0]]=[device_label[i]]
        else:
            days[date_time[0]][date_time[1].split(':')[0]].append(device_label[i])
            
print(days)
```

```python
hours=[]
count=[]
day=[]
for keys,items_mas in days.items():
    day.append(keys)
    temp_hour=[]
    temp_count=[]
    for sub_keys,sub_items in items_mas.items():
        counting=np.count_nonzero(sub_items)
        temp_hour.append(sub_keys)
        temp_count.append(counting)
    hours.append(temp_hour)
    count.append(temp_count)
```

```python
%matplotlib inline
print(len(days))
i=2
plt.plot(hours[i],count[i])
plt.title(day[i])
```
