```python
import pandas as pd
import keras
from keras.models import Sequential,Model
from keras.layers import Dense,Input,LSTM
from keras.utils import to_categorical
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('test_data.csv')
columns=data.columns

x=[]
for name in columns:
    x.append(data[name])
```

```python
label_encoder=LabelEncoder()
days_encoded=label_encoder.fit_transform(x[0])
```

```python
X=[]
for i in range(len(days_encoded)):
    temp=[]
    temp.append(float(days_encoded[i]))
    temp.append(float(x[1][i]))
    temp.append(float(x[2][i]))
    X.append(temp)
    
X=np.asarray(X)
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d_t' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d_t+%d' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
pre_window=2
post_window=2
time_series_Data=series_to_supervised(X,pre_window,post_window)
time_series_Data.drop(time_series_Data.columns[[6,9]], axis=1, inplace=True)
time_series_Data.head(5)
```

```python
# time_series_Data.drop(time_series_Data.columns[[6,9]], axis=1, inplace=True)
X=[]
Y=[]
for index,rows in time_series_Data.iterrows():
    temp_x=[]
    temp_y=[]
    for j in range(pre_window*len(columns)):
        temp_x.append(np.asarray(rows[j]))
    for k in range(pre_window*len(columns),len(rows)):
        temp_y.append(np.asarray(rows[k]))
    X.append(np.asarray(temp_x))
    Y.append(np.asarray(temp_y))
X=np.asarray(X)
Y=np.asarray(Y)
```

```python
master_x=X.reshape((X.shape[0],1,X.shape[1]))
master_y=Y.reshape((Y.shape[0],Y.shape[1]))
master_x[0]
```

```python
model = Sequential()
model.add(LSTM(50, input_shape=(master_x.shape[1], master_x.shape[2])))
model.add(Dense(4))
model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
model.fit(master_x,master_y,epochs=200,batch_size=4)
```

```python
test_inp=np.asarray([master_x[3]])
print(test_inp)
result=model.predict(test_inp)
print(result)
```
